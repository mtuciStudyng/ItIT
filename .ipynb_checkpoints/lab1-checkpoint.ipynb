```python
from random import randint as rnd

m = 23
n = 11
min_limit = -100
max_limit = 400

matrix = [
    [
        rnd(min_limit, max_limit) 
        for i in range(n)
    ]
    for _ in range(m)
]

print('\n'.join([str(el) for el in matrix]))
```

```python
# SELECT SORT
def select_sort(arr: list):
    size = len(arr)
    right_pos = size - 1
    
    while right_pos > 0:
        max_value, max_value_ind = float('-inf'), -1
        for k in range(right_pos + 1):
            if max_value < arr[k]:
                max_value = arr[k]
                max_value_ind = k
        arr[right_pos], arr[max_value_ind] = arr[max_value_ind], arr[right_pos]
        right_pos -= 1
    return arr
```

```python
# INSERT SORT
test_arr = [7,6,5,4,3,2,1]
def insert_sort(arr: list):
    size = len(arr)
    for k in range(size):
        el, i = arr[k], k - 1
        while i >= 0 and el < arr[i]:
            arr[i+1] = arr[i]
            i -= 1
        arr[i+1] = el
    return arr
```
